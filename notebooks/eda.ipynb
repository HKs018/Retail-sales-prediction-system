# Retail Sales EDA

This notebook performs beginner-friendly exploratory data analysis for the retail sales prediction project.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid')
```

```python
df = pd.read_csv('../sales_data.csv')
df.head()
```

```python
df.info()
df.describe()
```

```python
print('Missing values:')
print(df.isnull().sum())
print('\nDuplicate rows:', df.duplicated().sum())
```

```python
df = df.drop_duplicates()
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month

for column in ['Region', 'Category', 'Sub-Category', 'Product Name']:
    df[column] = df[column].fillna(df[column].mode()[0])

df['Profit'] = df['Profit'].fillna(df['Profit'].median())
df.head()
```

```python
monthly_sales = df.set_index('Order Date').resample('MS')['Sales'].sum()
monthly_sales.plot(figsize=(12, 5), marker='o', title='Monthly Sales Trend')
plt.ylabel('Sales')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='Region', y='Sales', estimator=sum)
plt.title('Sales by Region')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='Category', y='Sales', estimator=sum)
plt.title('Sales by Category')
plt.show()
```

```python
plt.figure(figsize=(7, 5))
numeric_columns = ['Sales', 'Profit', 'Quantity', 'Order Month', 'Order Year']
sns.heatmap(df[numeric_columns].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
